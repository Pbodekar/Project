Problem Statement:
Every year a lot of companies hire a number of employees. The companies invest time and money in training those employees, not just this but there are training programs within the companies for their existing employees as well. The aim of these programs is to increase the effectiveness of their employees. But where HR Analytics fit in this? and is it just about improving the performance of employees?

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
# reading file
file = 'C:\\Users\\praja\\Downloads\\ibm-hr-analytics-employee-attrition-performance\\Employee-Attrition.csv'
data = pd.read_csv(file)
data.head()
```

```python
data.columns
```

```python
# Checking missing values
data.isnull().sum()
```

```python
data.dtypes
```

```python
# create a DataFrame  with numeric variables
data_num = data[['Age','DailyRate','DistanceFromHome','Education','EmployeeNumber','EnvironmentSatisfaction','HourlyRate',
                 'JobInvolvement','JobLevel','JobSatisfaction','MonthlyIncome','MonthlyRate', 'NumCompaniesWorked',
                 'PercentSalaryHike', 'PerformanceRating','RelationshipSatisfaction', 'StockOptionLevel','TotalWorkingYears',
                 'TrainingTimesLastYear', 'WorkLifeBalance','YearsAtCompany', 'YearsInCurrentRole', 'YearsSinceLastPromotion',
                 'YearsWithCurrManager']].copy()
```

```python
plt.figure(figsize = (10,10), dpi = 100)
sns.heatmap(data_num.corr())
```

```python
# lets remove strongly ccorrelated variables
data_uc = data_num[['Age','DailyRate','DistanceFromHome','EnvironmentSatisfaction','HourlyRate','JobInvolvement',
                    'JobLevel','JobSatisfaction','RelationshipSatisfaction', 'StockOptionLevel','TrainingTimesLastYear']].copy()
```

```python
data_uc.head()
```

```python
# Categorical data
data_categorical = data[['Attrition', 'BusinessTravel','Department','EducationField','Gender','JobRole','MaritalStatus',
                       'Over18', 'OverTime']].copy()
data_categorical.head()
```

```python
Num_val = {'Yes':1, 'No':0}
data_categorical['Attrition'] = data_categorical["Attrition"].apply(lambda x: Num_val[x])
data_categorical.head()
```

```python
data_categorical = pd.get_dummies(data_categorical)
data_categorical.head()
```

```python
data_final = pd.concat([data_num, data_categorical], axis = 1)
data_final.head()
```

Model Building

```python
target = data_final['Attrition']
features = data_final.drop('Attrition', axis = 1)

# train and test split
x_train, x_test, y_train, y_test = train_test_split(features, target, test_size = 0.4, random_state = 10)
```

```python
model = RandomForestClassifier()
model.fit(x_train,y_train)
```

```python
# predictinng result for test
test_pred = model.predict(x_test)
accuracy_score(y_test, test_pred)
```

```python
feat_imp = pd.Series(model.feature_importances_, index = features.columns)
feat_imp = feat_imp.nlargest(20)
feat_imp.plot(kind = 'barh')
```
